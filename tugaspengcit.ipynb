```python
#Import Library
import cv2
import matplotlib.pyplot as plt

#1. Membaca Gambar
img = cv2.imread("pengcit.jpg")

#2. Konversi Gambar
#a. True Color (RGB untuk Matplotlib)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#b. Grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#c. Biner
ret, img_binary = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

#3. Tampilkan 3 Gambar
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title("True Color (RGB)")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(img_binary, cmap="gray")
plt.title("Biner")
plt.axis("off")

plt.tight_layout()
plt.show()

#4. Resize Gambar True Color ke 30x30
img_rgb_resized = cv2.resize(img_rgb, (30, 30))

#5. Cuplik Area 5x3 (tinggi=5, lebar=3)
pixel_matrix_5x3 = img_rgb_resized[0:5, 0:3]

#6. Cetak Matriks
print("Matriks Piksel 5x3 (dari gambar 30x30 True Color)")
print("Shape matriks:", pixel_matrix_5x3.shape, "(tinggi=5, lebar=3, RGB=3)")
print("\nNilai RGB setiap piksel:")
for i in range(5):
    for j in range(3):
        r, g, b = pixel_matrix_5x3[i, j]
        print(f"Piksel[{i},{j}]: RGB({r:3d}, {g:3d}, {b:3d})")

#7. Tampilkan Gambar Resize 30x30
plt.figure(figsize=(4, 4))
plt.imshow(img_rgb_resized)
plt.title("Gambar True Color (30x30)")
plt.axis("off")
plt.show()
```
